```python
%matplotlib inline
mpl.rcParams['figure.figsize'] = (20, 4)
mpl.rcParams['axes.grid'] = False

from scipy import stats
```

```python
action_steps = 83
freq = 15
simulation_steps = action_steps * freq
```

```python
def plot(x,y):
    _ = plt.plot(x,y)
    _ = plt.xlabel('steps')
    _ = plt.ylabel('demand (veh/hr)')
```

```python
def generate_demand(max_demand=10000, periods=2, steps=100, simulation_freq=15):
    simulation_steps = steps * simulation_freq
    x = np.linspace(0, periods * 2 * np.pi, simulation_steps)
    y1 = ((np.sin(x) + 1) / 2) * max_demand
    y2 = max_demand - y1
    
    return y1, y2
```

```python
y1, y2 = generate_demand(periods=1)
x = np.arange(len(y1))
_ = plt.plot(x,y1)
_ = plt.plot(x,y2)
_ = plt.xlabel('steps')
_ = plt.ylabel('demand (veh/hr)')
_ = plt.show()
```

```python
AMPLITUDE = np.random.normal(10000, 2000)
PERIODS = np.abs(np.random.normal(2,1))
x = np.linspace(0, PERIODS * (2*np.pi), simulation_steps)
y1 = np.random.normal(0.5, 0.1) * (np.sin(x) + np.random.normal(2, 0.2)) * AMPLITUDE
y2 = np.min(y1) + (np.max(y1) - y1)

plot(x,y1)
plot(x,y2)
```
