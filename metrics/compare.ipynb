```python
import warnings; warnings.filterwarnings('ignore')

import os
import sys

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
mpl.rcParams['figure.figsize'] = (18, 3)
mpl.rcParams['axes.grid'] = False

from pathlib import Path
base = Path('.')
```

```python
path1 = base / 'exp1' / 'baseline1.csv'
df1 = pd.read_csv(path1)

path2 = base / 'exp1' / 'baseline2.csv'
df2 = pd.read_csv(path2)
```

```python
a = 0.2
metric = 'total_reward'

for df, col in zip([df1, df2], ['red', 'blue']):
    
    for ep in df.episode.unique():
        
        df_ep = df.loc[df.episode == ep]
        x = df_ep.step.values
        
        y = df_ep[metric]
        
        _ = plt.plot(x, y, c=col, alpha=a)
    
    df_mean = df.groupby('step', as_index=False).mean()
    
    x = df_mean.step.values
    y = df_mean[metric]
    
    _ = plt.plot(x, y, c=col)
```
